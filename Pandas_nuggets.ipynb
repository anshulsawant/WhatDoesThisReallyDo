```python
import pandas as pd
import numpy as np
```

```python
#@title Expanding windows
df = pd.DataFrame(np.array([[1, 1, 2, 2], [1, 2, 1, 2], [5, 6, 8, 7]]).transpose(), columns=['A', 'B', 3])
print(df)
df.sort_values(by=['A', 'B'], inplace=True)
df['expanding_sum'] = df.groupby(['A'])[3].expanding().min().reset_index(drop=True)
df
```
